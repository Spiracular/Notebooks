```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#Default plotting w/ 1 array: assumes it's y-axis,
#generates evenly-spaced x values starting at 0
plt.plot([1,2,3])
plt.show()
plt.plot([0,1,4,9,16,25])
plt.show()
```

```python
x = np.linspace(0, 6)
plt.ylabel('Axis Y')
plt.xlabel ('Axis X')
plt.plot(sin(x), x, 'ro')
plt.show()
```

```python
#Subplot formation: (#xrows, #yrows, plot#)
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
x2 = np.linspace(0.0, 5.0)
y2 = np.cos(2*np.pi*x2)
x3 = np.linspace(0.0, 5.0)
y3 = np.exp(-x1)

plt.subplot(3,1,1)
plt.plot(x1, y1, 'yo')
plt.title('Components of Oscillation')
plt.ylabel('Damped')

plt.subplot(3,1,2)
plt.plot(x2, y2, 'ro-')
plt.xlabel('time(s)')
plt.ylabel('Undamped')

plt.subplot(3,1,3)
plt.plot(x3, y3, 'b.-')
plt.ylabel('Dampening')

plt.show()
```

```python
#4 subplots
x = np.linspace(0, 4)
y = np.linspace (0, 4)

plt.subplot(2,2,1)
plt.plot(x, y)

plt.subplot(2,2,2)
plt.plot(x, -y)

plt.subplot(2,2,3)
plt.plot(-x, y)

plt.subplot(2,2,4)
plt.plot(-x, -y)
```

```python
#Pie Chart

labels = 'Apple', 'Cherry', 'Pumpkin', 'Meat'
sizes = [45, 35, 15, 5]
colors = ['yellow', 'red', 'orange', 'brown']
explode = (0, 0, 0, 0.2)
plt.pie(sizes, explode = explode, labels = labels, colors = colors,
        autopct = '%1.1f%%', shadow = True, startangle = 0)
#Seems to turn it from slanted oval to a circular straightforward view
plt.axis('equal')
plt.show()
```
