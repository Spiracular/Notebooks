## Messing with color palettes

[Contour code](http://stackoverflow.com/questions/31095772/using-seaborn-and-contourf-how-can-i-plot-gridlines)

[MatPlotLib example](http://jakevdp.github.io/mpl_tutorial/tutorial_pages/tut3.html)

[Seaborn color palette details](https://stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html)

Once you get this to work, your next goal may be to make something work with Lat, Long, and stormdata?

```python
dx=0.05
x=np.arange(0,5+dx,dx)
y=x
X,Y = np.meshgrid(x,y)
Z = np.sin(X)**10+np.cos(10+Y*Y)*np.cos(X)

nbins=10
levels=matplotlib.ticker.MaxNLocator(nbins=nbins).tick_values(Z.min(),Z.max())

plt.figure()
plt.contourf(x,y,Z,levels=levels)
plt.colorbar()
plt.grid('on')

sns.set_context("notebook")
sns.set_style("whitegrid")
sns.color_palette("Blues")

plt.figure()
plt.contourf(x, y, Z, levels=levels, zorder=0)
plt.colorbar()
plt.grid('on')
plt.show()
```
